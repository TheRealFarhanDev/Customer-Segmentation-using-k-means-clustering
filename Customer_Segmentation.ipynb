
# Customer Segmentation using K-Means Clustering

## Project Objective
The objective of this project is to group customers into meaningful segments based on:
- Annual Income
- Spending Score

This helps businesses:
- Identify high-value customers
- Improve marketing strategies
- Increase customer satisfaction



# Step 1: Import Libraries


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```


# Step 2: Load Dataset


```

dataset = pd.read_csv('Mall_Customers.csv')

dataset.head()
```


# Step 3: Dataset Information


```

dataset.info()
```

```

dataset.describe()
```


# Step 4: Feature Selection

We use:
- Annual Income
- Spending Score


```

X = dataset[['Annual Income (k$)', 'Spending Score (1-100)']].values

X[:5]
```


# Step 5: Elbow Method

The Elbow Method helps find the optimal number of clusters.


```

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```


# Step 6: Apply K-Means Clustering

From the Elbow graph, we choose:
- K = 5


```

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)

y_kmeans = kmeans.fit_predict(X)

y_kmeans
```


# Step 7: Visualize Customer Segments


```

plt.figure(figsize=(10,7))

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=50, label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=50, label='Cluster 5')

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=300,
    c='yellow',
    label='Centroids'
)

plt.title('Customer Segmentation')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```


# Step 8: Cluster Interpretation

### Cluster Analysis
- Cluster 1 → High income, high spending customers
- Cluster 2 → Low income, low spending customers
- Cluster 3 → Moderate customers
- Cluster 4 → Careful spenders
- Cluster 5 → Target customers

Businesses can use this information for targeted marketing strategies.

Customer Segmentation using K-Means Clustering helps businesses:
